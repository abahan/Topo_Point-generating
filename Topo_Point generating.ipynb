```python
class Point:
    def __init__(self,x_init,y_init,z_init = 0):
        self.x = x_init
        self.y = y_init
        self.z = z_init  # Z in meters whereas x,y in mm Survey points
       
       
        

    def shift(self, x, y, z=0):
        self.x += x
        self.y += y
        self.z += z

    def __repr__(self):
        return "".join(["Point(","," ,str(self.x), ",", str(self.y),  ")"])
    
    def distance(a, b):
        return math.sqrt((a.x-b.x)**2+(a.y-b.y)**2)
```

```python
# Point list to DF database
def point_list_to_df(p_list):
    p_l={}

    p_l['X']=[]
    p_l['Y']=[]
    p_l['Z']=[]
    
    for pt in p_list:

        p_l['X'].append(pt.x)
        p_l['Y'].append(pt.y)
        p_l['Z'].append(pt.z)

    
    df_points=pd.DataFrame(p_l)
        
    return df_points 
```
